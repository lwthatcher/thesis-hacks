```python
# --------------------------------------------------
#
# This one charts the graph inline
#
# --------------------------------------------------

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

FILE = 'pills-s0.npy'

X = np.load(FILE)

print(X.shape)

plt.plot(X)
plt.show()
```
