```python
# --------------------------------------------------
#
# The overall spectral energy for the entire data set
#
# --------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

FILE = 'data/pills-s0.npy'
x = np.load(FILE)
print(x.shape)

f, p = signal.welch(x, 10e3, axis=0)

print(f.shape, p.shape)

plt.plot(p)
plt.show()
```

```python
# --------------------------------------------------
#
# The spectral energy for a 'quiet' region
#
# --------------------------------------------------

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

FILE = 'data/pills-s0.npy'
x = np.load(FILE)
print(x.shape)

f, p = signal.welch(x[2250:2400, :], 10e3, axis=0)

print('f', f.shape)

print('Pxx', p.shape)

plt.plot(p)
plt.show()
```
